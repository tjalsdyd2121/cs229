#numpy, matplotlib and some useful packages

#### numpy array

```python
import numpy as np

data = np.array([[1,2,3],[4,5,6]])
n,d = data.shape
print(data)
print(n,d)
```

```python
import numpy as np

zero_vector = np.zeros(5)
print(zero_vector)

# useful operation for numpy array
print(1-zero_vector)

# implement inner product
print(zero_vector.T) # T stands for transpose
print((1-zero_vector).T.dot(3-zero_vector))
```

```python
import numpy as np

# norm
vec =  1 - np.zeros(5)
# sum of every elements, norm 1
print(np.linalg.norm(vec , ord=1))
# Euclidean distance, norm 2
print(np.linalg.norm(vec , ord=2))
```

##### numpy inverse matrix

```python
import numpy as np

#inverse matrix
mat = np.array([[2,0],[0,2]])
print(mat)
print(np.linalg.inv(mat))
```

##### numpy array sum

```python
import numpy as np
#useful summation on data set

a = np.array([[1, 0, 3],
             [2, 5, 4],
             [3, 6, 7]])
print(np.sum(a))
# sum as row
print(np.sum(a,axis=0))
# sum as column
print(np.sum(a,axis=1))
# keeping the dimensions
print(np.sum(a,axis=0,keepdims=True))
print(np.sum(a,axis=1,keepdims=True))
```

##### numpy array boolean indexing

```python
import numpy as np

a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 1, 3])
# For only numpy array.
c = [5,6,7,8,9]
d = [1,2,3,4]
one = a[b==1]
# for ordinary array, the statement "d==1" is False.
# so d==1 -> 0.
wrong = c[d==1]
print(one, wrong)
```
